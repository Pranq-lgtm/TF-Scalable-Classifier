```python
import tensorflow as tf

scalar = tf.constant(5)
vector = tf.constant([1,2,3])
matrix = tf.constant([[1,2],[3,4]])
tensor_3d = tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])

rank_scalar = tf.rank(scalar)
rank_vector = tf.rank(vector)
rank_matrix = tf.rank(matrix)
rank_3d = tf.rank(tensor_3d)

print("Rank of Scalar:",rank_scalar.numpy())
print("Rank of Vector:",rank_vector.numpy())
print("Rank of Matrix:",rank_matrix.numpy())
print("Rank of 3D Tensor:",rank_3d.numpy())
```

```python
t2 = tf.constant([[0,1,2],
                  [3,4,5],
                  [6,7,8]])

print(t2)

first_row = tf.slice(t2,[0,0],[1,tf.shape(t2)[1]])
print("First_row:\n",first_row)

second_row = tf.slice(t2,[1,0],[1,tf.shape(t2)[1]])
print("\nSecond Row:\n",second_row)

second_third_rows = tf.slice(t2,[1,0],[2,tf.shape(t2)[1]])
print("\nSecond & Third Rows:\n",second_third_rows)

second_column=tf.slice(t2,[0,1],[tf.shape(t2)[0],1])
print("\nSecond Column:\n",second_column)

first_column = tf.slice(t2,[0,0],[tf.shape(t2)[0],1])
print("\nFirst Column:\n",first_column)

second_third_columns = tf.slice(t2,[0,1],[tf.shape(t2)[0],2])
print("\nSecond & Third Columns:\n",second_third_columns)
```

```python
import tensorflow as tf

t2 = tf.constant([[0,1,2],
                  [3,4,5],
                  [6,7,8]])

first_row = t2[0,:]
print("First Row:\n",first_row)

second_row = t2[1,:]
print("\nSecond Row:\n",second_row)

second_third_rows = t2[1:3,:]
print("\nSecond & Third Rows:\n",second_third_rows)

first_column = t2[:,0]
print("\nFirst Column:\n",first_column)

first_second_columns = t2[:,0:2]
print("\nFirst & Second Columns:\n",first_second_columns)
```

```python
import tensorflow as tf

t1 = tf.constant([0,1,2,3,4,5,6,7])

favorites = tf.gather(t1, indices=[0,3,6])

print(favorites)


t2 = tf.constant([[0,1,2],
                  [3,4,5],
                  [6,7,8]])

Indices = tf.constant([1,0])
r = tf.gather(t2, indices=Indices, axis=0)
print("Along Row\n",r)

c = tf.gather(t2, indices=Indices, axis=1)
print("\nAlong Columns\n",c)

```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
@tf.function
def quadratic(x):
  return x**2+2*x+1
x_vals=tf.range(-10,11,1,dtype=tf.float32)
y_vals=quadratic(x_vals)
plt.plot(x_vals.numpy(),y_vals.numpy(),marker='o')
plt.title("Quadratic Function:Y=x^2+2x+1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid (True)
plt.show()
```
